```python
import pandas as pd
url='/content/classification.csv'
data=pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation=data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot=True)
```

```python
sns.pairplot(data)
```

```python
#Bc 1
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['interest']), y = data['success'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la:',model.coef_)
```

```python
model.score(X=pd.DataFrame(data['interest']), y = data['success'])
```

```python
y_pred = model.predict(X=pd.DataFrame(data['interest']))
```

```python
data.plot(kind='scatter',x='interest', y = 'success' , figsize=(9,9), color= 'green')
plt.plot(data['interest'],y_pred,color='black')
```

```python
#Bc 2
import numpy as np
model = linear_model.LinearRegression()
X= pd.DataFrame([data['interest'],data['interest']**2]).T
y=data['success']
model.fit(X,y)
print('Intercept cua mo hinh la:',model.intercept_)
print('Coeffient cua mo hinh la:',model.coef_)
print('Do chinh xac cua mo hinh la:',model.score(X,y))
range=np.arange(0,101,1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='interest', y = 'success' , figsize=(9,9),color= 'green')
plt.plot(range, y_pred, color='black')
```
